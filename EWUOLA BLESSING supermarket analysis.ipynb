# DATA CLEANING

```python
#make a copy of the working data

df= data.copy()
```

```python
#correct column names

#remove leading or trailing spaces
df.columns = df.columns.str.strip()

#change column names to lower cases
df.columns = df.columns.str.lower()

#replace the spaces in between names to underscore
df.columns = df.columns.str.replace(' ', '_', regex=False)

#remove the extra space in supermarket size column name
df.columns = df.columns.str.replace('supermarket__size','supermarket_size',regex = False)
```

```python
#show the corrected column names

df.columns
```

THE PRODUCT WEIGHT COLUMN

```python
#fill in the null product weights with the weights from similar identifiers

df = df.sort_values('product_identifier')
df['product_weight'] = df['product_weight'].fillna(method = 'ffill')
df['product_weight'] = df['product_weight'].fillna(method = 'bfill')
```

```python
#confirm  that null values have been filled

df['product_weight'].isnull().sum()
```

THE PRODUCT FAT CONTENT COLUMN

```python
#show the dirty fat content column

data['Product Fat Content'].unique()
```

```python
# clean the product fat content column

content_replace ={'L.*' : 'Low Fat', 'N.*' : 'Normal Fat', 'U.*': 'Ultralow Fat'}

# Replace the old values with the new ones from the dictionary
df['product_fat_content'] = df['product_fat_content'].replace(content_replace, regex=True)
```

```python
#display the corrected values

df['product_fat_content'].unique()
```

PRODUCT TYPE COLUMN

```python
#show the dirty product type column

data['Product Type'].unique()
```

```python
product_replace ={'Snack.*' : 'Snacks', 'Health.*' : 'Health and Hygiene', 'Fruits and Veg.*': 'Fruits and Vegetables',
                  'Hard Drinks': 'Alcohol', 'Baking.*': 'Baking Items', 'Can.*': 'Canned Foods', 'Starch.*': 'Starchy Foods',
                 'Bread.*' : 'Bread'}

# Replace the values with the new names
df['product_type'] = df['product_type'].replace(product_replace, regex=True)
```

```python
#show corrected product types

df['product_type'].unique()
```

SUPERMARKET OPENING YEAR

```python
#show the dirty opening column

data['Supermarket Opening Year'].unique()
```

```python
year_replace ={'.*009':'2009', '.*004':'2004', '.*006':'2006'}

# Replace the values with the new names
df['supermarket_opening_year'] = df['supermarket_opening_year'].replace(year_replace, regex=True)
```

```python
#show corrected year values

df['supermarket_opening_year'].unique()
```

SUPERMARKET SIZE

```python
data['Supermarket  Size'].unique() 
```

```python
#replace wrong spellings with the correct ones
size_replace = {'Med.*|med.*': 'Medium','Sma.*|sma.*': 'Small', 'H.*':'Large'}

df['supermarket_size'] = df['supermarket_size'].replace(size_replace, regex=True)
```

```python
#fill in the null supermarket sizes

#find the mode of the supermarket class
size_mode = df['supermarket_size'].mode()[0]

#fill in null values with the mode of supermarket size
df['supermarket_size'] = df['supermarket_size'].fillna(size_mode)
df['supermarket_size'].value_counts()
```

```python
# confirm that there are no null values anymore 

df.isnull().sum()
```

```python
#round up the numeric values of some columns 

round_data= ['product_weight','product_price','product_supermarket_sales']
df[round_data] = df[round_data].round(2)
df[round_data]
```

# VISUALIZATION

```python
# to show how much fatty foods are being sold at the supermarkets


sns.countplot(data=df, x="product_fat_content", )
plt.title('PRODUCT FAT CONTENTS')
plt.xlabel('Fat Content')
plt.ylabel('Count of Sales')
sns.despine()

#the supermarket sells more lowfat foods
```

The supermarkets sold more low fat products than the ultralow and normal fat items

```python
# to show types of products being sold at the supermarkets


sns.countplot(data=df, y="product_type")
plt.title('Supermarket Products')
plt.ylabel('Products')
plt.xlabel('Quantity')
sns.despine()

#the supermarket sells more lowfat foods
```

The  fruits and vegetables is the most sold product, closely followed by snacks. The seafood were the least sold products.

```python
# using a pie chart to show the size of supermarkets

label= df['supermarket_size'].unique()
colour=['gold','lightcoral','lightskyblue']
plt.pie(df['supermarket_size'].value_counts(), labels=label, colors=colour)
plt.title('Supermarket sizes')

#show plot
plt.show()
```

Most of the supermarkets are medium sized. The large sized supermarkets are the fewest.

```python
# to show the supermarket that made the highest sale

sns.barplot(data=df , y='supermarket_identifier' , x='product_supermarket_sales', ci=None)
plt.title('PRODUCT SALES')
plt.xlabel('Sales')
plt.ylabel('Products')
sns.despine()
```

The CHUKWUDI027 supermarkets made the highest sales

```python
#otal sales made from each product type from the highest

new= df.groupby('product_type')['product_supermarket_sales'].sum().reset_index()
sort= new.sort_values(by='product_supermarket_sales', ascending= False)
```

```python
sns.set_style('white')
sns.color_palette('rocket', as_cmap=True)
sns.barplot(y=sort['product_type'],x=sort['product_supermarket_sales'])
plt.title('PRODUCT SALES')
plt.xlabel('Sales')
plt.ylabel('Products')
sns.despine()
```

The fruits and vegetables had the most sale amount while the seafood has the least sale amount

```python
# To show the sales of each products according to the supermarket size
sns.set_style('white')
sns.set_palette('Set1')

plt.figure(figsize=(10,40))

sns.barplot(data=df, x='product_supermarket_sales', y='product_type', hue='supermarket_size' , ci=None)
plt.xlabel('sales')
plt.ylabel('Products')
plt.title("Sales of products according to supermarket size")

sns.despine()
```
